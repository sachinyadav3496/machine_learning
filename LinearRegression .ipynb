```python
from sklearn import datasets
# sci-kit learn 
```

```python
# boston house data set 
```

```python
house = datasets.load_boston()
```

```python
print(type(house))
```

```python
print(house.DESCR)
```

```python
print(house.feature_names)
```

```python
features = pd.DataFrame(
    data = house.data,
    columns= house.feature_names
)
```

```python
features.head()
```

```python
target = house.target
```

```python
target[:5]
```

    X_train --> traing features data 
    X_test ---> testing features data
    y_train ---> traing target data 
    y_test  ---> traget testing data 

```python
from  sklearn.model_selection import train_test_split
```

```python
# train_test_split(features,target,test_size) --> xtrain,xtest,ytrain,test
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=.25)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
# let's train your model 

model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
predicted = model.predict(X_test)
```

```python
actual = y_test
```

```python
predicted[:5]
```

```python
actual[:5]
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_absolute_error(actual,predicted)
```

```python
for i in X_test.columns : 
    plt.title(f"According to {i}")
    plt.scatter(X_test[i],predicted,c="r",label="Prediction")
    plt.scatter(X_test[i],actual,c="g",label="Actual")
    plt.legend()
    plt.show()
```
