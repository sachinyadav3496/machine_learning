```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=3,
                 cluster_std=1,random_state=101)
```

```python
print(type(data))
```

```python
len(data)
```

```python
data[0][:5]
```

```python
data[1][:5]
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(10,4))
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],
           cmap='coolwarm')
plt.show()
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(data[0])
```

```python
clf.cluster_centers_
```

```python
clf.labels_
```

```python
fig,(ax1,ax2)  = plt.subplots(2)
ax1.scatter(data[0][:,0],data[0][:,1],c=clf.labels_,cmap='coolwarm',
           label="Prediction")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],
           cmap='coolwarm',label="Actual")
plt.legend()
plt.show()
```
